```python
import pandas as pd

nt = pd.read_csv(r"/content/nutrients_csvfile(3).xls")
print(nt)
```

```python
nt.dtypes
```

```python
nt.replace("t",0,inplace=True)
nt=nt.replace("t'",0)
nt.replace(",","", regex=True, inplace= True)
nt['Fiber']=nt['Fiber'].replace("a","", regex=True, inplace= True)
nt['Calories'][91]=(8+44)/2
```

```python
nt['Grams']=pd.to_numeric(nt['Grams'])
nt['Calories']=pd.to_numeric(nt['Calories'])
nt['Protein']=pd.to_numeric(nt['Protein'])
nt['Fat']=pd.to_numeric(nt['Fat'])
nt['Sat.Fat']=pd.to_numeric(nt['Sat.Fat'])
nt['Fiber']=pd.to_numeric(nt['Fiber'])
nt['Carbs']=pd.to_numeric(nt['Carbs'])
```

```python
nt.dtypes
```
